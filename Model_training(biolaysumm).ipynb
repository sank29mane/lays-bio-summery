```python
# Load model directly
#import torch
#import torch.nn as nn
#import bitsandbytes as bnb
from transformers import AutoModelForCausalLM, AutoTokenizer
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2", torch_dtype="auto", load_in_8bit=True, trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2", trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
```

```python
from peft import LoraConfig, get_peft_model

config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=['Wqkv','out_proj'],
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, config)
```

```python
def formatting_func(example):
  output_texts = []
  for i in range(len(example['article'])):
      text = f''' Summarise the following article in simple/lay terms
      ### article: {example['article'][i]}\n
      ### lay_summary: {example['lay_summary'][i]}

      '''
      output_texts.append(text)
  return output_texts
```

```python
# Define training arguments
from transformers import trainer, TrainingArguments
args = TrainingArguments(
    output_dir="/content/drive/MyDrive/Dissertation1/outputft5b3",
    evaluation_strategy="steps",
    learning_rate=2e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    fp16=True,
    num_train_epochs=1,
    warmup_steps=500,
    max_grad_norm=1,
    weight_decay=0.01,
    logging_dir="/content/drive/MyDrive/Dissertation1/logsft5b3",
    logging_steps=20,
    eval_steps=30,
    gradient_accumulation_steps=60,

)
```

```python
from transformers import Adafactor

optimizer = Adafactor(
    model.parameters(),
    lr=2e-5,
    clip_threshold=1.0,
    weight_decay=0.01,
    relative_step=False,

)
```

```python
train_dataset = train_dataset['train']
trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    args=args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    formatting_func=formatting_func,
    max_seq_length=512,
    #packing=True,
    peft_config= config,
    optimizers=(optimizer, None),


)
```

```python
trainer.train()
```
